```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
train_dir = os.path.join('./','train/')
train_images = os.path.join(train_dir, 'images')
train_labels = os.path.join(train_dir, 'labels')
```

```python
label_files = os.listdir(train_labels)
image_files = os.listdir(train_images)
```

```python
def extract_labels(train_labels, label_filename):
    '''
    Input: Training labels folder and label filename
    Output: Dictionary of classlabels and bounding box for respective image
    '''
    label_files = os.path.join(train_labels, label_filename)
#     print(label_files)
    class_label, bb_label = list(), list()
    class_dict, bb_dict = dict(), dict()
    with open(label_files) as f:
        s = f.readlines()
        for i in s:
            i = i.replace('\n','')
            characters = i.split(' ')
        class_label.append(characters[0])
        bb_label.append(characters[1:])
#     print(class_label)
#     print(bb_label)
    class_dict[label_filename] = class_label
    bb_dict[label_filename] = bb_label
    return class_dict, bb_dict

class_list, bb_list = list(), list() 
class_pd = pd.DataFrame(columns=['image', 'class'])
bb_pd = pd.DataFrame()
for i in label_files:
    class_dict, bb_dict = extract_labels(train_labels, i)
#     print(bb_dict)
    temp_dict = {'image':os.path.join(train_labels, i), 'class':class_dict[i]}
    class_pd = class_pd.append(pd.DataFrame(temp_dict), ignore_index=True)
```
